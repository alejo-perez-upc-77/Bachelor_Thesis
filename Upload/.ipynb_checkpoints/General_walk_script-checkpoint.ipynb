```python
import os
import pandas as pd
import matplotlib as plt
plt.rcParams.update({'font.size': 20})
```

```python
dic = {'NORMAL_LYMPHOCYTE':[],
      'ATYPICAL_LYMPHOCYTE':[],
      'VARIANT_LYMPHOCYTE':[],
      'BLAST':[]
     }


def print_dirs_recursively(root_dir):
    root_dir = os.path.abspath(root_dir)

    for item in os.listdir(root_dir):
        item_full_path = os.path.join(root_dir, item)
        if os.path.isdir(item_full_path):
            print_dirs_recursively(item_full_path)
        else: 
            if '/Linfos_normales/' in item_full_path:
                dic['NORMAL_LYMPHOCYTE'].append(item_full_path)
            if '/ATYPICAL_LYMPHOCYTE/' in item_full_path:
                dic['ATYPICAL_LYMPHOCYTE'].append(item_full_path)
            if '/VARIANT_LYMPHOCYTE/' in item_full_path:
                dic['VARIANT_LYMPHOCYTE'].append(item_full_path)
            if '/BLAST/' in item_full_path:
                dic['BLAST'].append(item_full_path)

print_dirs_recursively('/shared/Clinic')
```

```python
df_Abs = pd.DataFrame.from_dict(dict(zip(list(dic.keys()),[[len(dic[i])]for i in dic.keys()])), orient='index')
df_Abs.rename({0:'Número de Imágenes en el conjunto absoluto'}, inplace=True, axis='columns' )
df_Abs
```

```python
df_Abs.plot.bar(rot=0, figsize=(20,10));
```

```python
import re
def catcher2(x): 
    label = re.findall('([a-zA-Z]+)/[\d]|([a-zA-Z]+_[a-zA-Z]+)/[\d]', x)[0]
    if label[0] == '': return label[1]
    else: return label[0]

typ_2 = {'LZME':0,
         'LF':0,
         'LCM_INDOLENTE':0,
         'LCM_BLASTICO':0,
         'LCM_TIPICO':0,
         'LLC':0}

for i in dic['ATYPICAL_LYMPHOCYTE']:
    if '.jpg' in i:
        label = catcher2(i)
        if label in typ_2:
            typ_2[label] += 1


```

```python
df_Atypical = pd.DataFrame(pd.Series(typ_2)).rename({0:'N en la Categoría ATYPICAL_LYMPHOCYTE'},axis=1)
```

```python
df_Atypical.plot.bar(rot=0, figsize=(20,10));
```
