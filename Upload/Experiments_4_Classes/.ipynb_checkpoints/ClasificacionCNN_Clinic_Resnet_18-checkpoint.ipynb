```python
import fastai
import torch
from fastai.vision import *
from shutil import copyfile
import random
import pandas as pd
import numpy as np
```

```python
dic = {'NORMAL_LYMPHOCYTE':[],
      'ATYPICAL_LYMPHOCYTE':[],
      'VARIANT_LYMPHOCYTE':[],
      'BLAST':[]
     }
```

```python
import os

def print_dirs_recursively(root_dir):
    root_dir = os.path.abspath(root_dir)

    for item in os.listdir(root_dir):
        item_full_path = os.path.join(root_dir, item)
        if os.path.isdir(item_full_path):
            print_dirs_recursively(item_full_path)
        else: 
            if '/Linfos_normales/' in item_full_path:
                dic['NORMAL_LYMPHOCYTE'].append(item_full_path)
            if '/ATYPICAL_LYMPHOCYTE/' in item_full_path:
                dic['ATYPICAL_LYMPHOCYTE'].append(item_full_path)
            if '/VARIANT_LYMPHOCYTE/' in item_full_path:
                dic['VARIANT_LYMPHOCYTE'].append(item_full_path)
            if '/BLAST/' in item_full_path:
                dic['BLAST'].append(item_full_path)

print_dirs_recursively('/shared/Clinic')
```

```python
for item in dic.values():
     random.shuffle(item)
```

```python
def catcher(x): return re.findall('/([a-zA-Z]+)_+[\d]', x)[0]

typ = {'NORMAL_LYMPHOCYTE':{},
      'ATYPICAL_LYMPHOCYTE':{},
      'VARIANT_LYMPHOCYTE':{},
      'BLAST':{}
     }
for name in dic.keys():
    for path in dic[name]:
        if '.jpg' in path:
            label = catcher(path)
            if label in typ[name]:
                typ[name][label] += 1

            else: 

                typ[name][label] = 1


typ
```

Creamos una FocalLoss porque funciona bien con un dataset desbalanceado

```python
class FocalLoss(nn.Module):
    def __init__(self, gamma=2):
        super().__init__()
        self.gamma = gamma

    def forward(self, logit, target):
        target = target.float()
        max_val = (-logit).clamp(min=0)
        loss = logit - logit * target + max_val + \
               ((-max_val).exp() + (-logit - max_val).exp()).log()

        invprobs = F.logsigmoid(-logit * (target * 2.0 - 1.0))
        loss = (invprobs * self.gamma).exp() * loss
        if len(loss.size())==2:
            loss = loss.sum(dim=1)
        return loss.mean()
```

```python
learner_focal = cnn_learner(data, models.resnet18, metrics=[error_rate,accuracy], pretrained='imagenet')
learner_focal.loss_fn = FocalLoss()
```

```python
learner_focal.lr_find()
learner_focal.recorder.plot()
```

```python
learner_focal.fit_one_cycle(8)
```

```python
learner_focal.fit_one_cycle(4)
```

```python
learner_focal.save('stage-1-18-Clinic_Focal')

learner_focal.unfreeze()
learner_focal.fit_one_cycle(4, max_lr=slice(1e-6,1e-4))
```

```python
learner_focal.save('stage-2-18-Clinic_Focal')

interp = ClassificationInterpretation.from_learner(learner_focal)

interp.most_confused(min_val=2)
```
